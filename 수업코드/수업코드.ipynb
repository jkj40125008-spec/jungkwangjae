```python
data = {
    '나이':[22,35,14,55,42]
}

df = pd.DataFrame(data)

def categorize(data):
    if 0 < data <= 10:
        return "10대"
    elif 10 < data <= 20:
        return "20대"
    elif 20 < data <= 30:
        return "30대"
    else:
        return "30대 이상"

df['검사나이'] = df['나이'].apply(categorize)

print(df)
```
