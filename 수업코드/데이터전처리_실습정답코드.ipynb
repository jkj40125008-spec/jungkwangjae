```python
import numpy as np
import pandas as pd
```

```python
lactate = pd.read_csv("data/sample_1.csv", encoding="utf-8")
```

```python
#선택하려고 하는 변수
data_list=['연구번호', '입실일자', "입실시각",'검사코드NM',"검사결과값(숫자)","기준치상한값", 
            "기준치하한값", "검사나이", "접수일자", "접수시각","퇴실일자","퇴실시각","환자생년월","성별코드"]
```

```python
lactate=lactate[data_list]
```

```python
lactate["환자ID"]=lactate["연구번호"].astype("str")+lactate["입실일자"].astype("str")+lactate["환자생년월"].astype("str")+lactate["퇴실시각"].astype("str")
```

```python
lactate["환자ID"]=lactate["환자ID"].str[0:3]+lactate["환자ID"].str[5:10]
```

```python
df=lactate[(lactate["검사코드NM"]=="Lactic acid (Qn)[EM],Artery blood")]
df=lactate.dropna()
```

```python
df["입실일자"]=pd.to_datetime(df["입실일자"], format='%Y%m%d', errors="coerce")
df["퇴실일자"]=pd.to_datetime(df["퇴실일자"], format='%Y%m%d', errors="coerce")
df["접수일자"]=pd.to_datetime(df["접수일자"], format='%Y%m%d', errors="coerce")
```

```python
df["퇴실시각"]=df["퇴실시각"].astype(str)
df["퇴실시각"]=df["퇴실시각"].str.rjust(4, fillchar="0")
```

```python
df["입실시각"]=df["입실시각"].astype(int)
df["입실시각"]=df["입실시각"].astype(str)
df["입실시각"]=df["입실시각"].str.rjust(4, fillchar="0")
```

```python
df["접수시각"]=df["접수시각"].astype(int)
df["접수시각"]=df["접수시각"].astype(str)
df["접수시각"]=df["접수시각"].str.rjust(4, fillchar="0")
```

```python
df["입실시각"]=pd.to_datetime(df["입실시각"], format='%H%M', errors="coerce").dt.time
df["퇴실시각"]=pd.to_datetime(df["퇴실시각"], format='%H%M', errors="coerce").dt.time
df["접수시각"]=pd.to_datetime(df["접수시각"], format='%H%M', errors="coerce").dt.time
```

```python
df["입원기간"]=df["퇴실일자"]-df["입실일자"]
```

```python
def categorize_age(age):
    if 0 < age <= 19:
        return "10대"
    elif 19 < age <= 29:
        return "20대"
    elif 29 < age <= 39:
        return "30대"
    elif 39 < age <= 49:
        return "40대"
    elif 49 < age <= 59:
        return "50대"
    elif 59 < age <= 69:
        return "60대"
    elif 69 < age <= 79:
        return "70대"
    elif 79 < age < 89:
        return "80세 "
    elif age >= 80:
        return "80세 이상"
    else:
        return "Under 10"
```

```python
df['검사나이'] = df['검사나이'].apply(categorize_age)
```

```python
df.to_csv('data/t1.csv', index=False, encoding='cp949')
```
